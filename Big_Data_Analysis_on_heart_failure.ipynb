## Checking for Outliers for RestingBP

```python
# 1. IQR method

# Create a function to return index of outliers
def indicies_of_outliers(x):
    q1, q3 = np.percentile(x, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    return np.where((x > upper_bound) | (x < lower_bound))
# # Run function
indicies_of_outliers(new_data['RestingBP'])
```

```python
# 2. Boxplot Visualization method

import matplotlib.pyplot as plt

# Create a box plot to visualize the data
plt.boxplot(new_data['RestingBP'])
plt.title('Box Plot of RestingBP')
plt.show()
```

```python
3. #Tukey's method

# Calculate the IQR for RestingBP
Q1 = new_data['RestingBP'].quantile(0.25)
Q3 = new_data['RestingBP'].quantile(0.75)
IQR = Q3 - Q1

# Define Tukey's Fences
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Find outliers
outliers = new_data[(new_data['RestingBP'] < lower_fence) | (new_data['RestingBP'] > upper_fence)]

# print("DataFrame:")
# print(new_data)
print("\nOutliers based on Tukey's Fences:")
print(outliers)
```

# Transforming the Outliers

### Transforming the Outliers in RestingBP column

```python
# We can transform the feature using the yeojohnson feature to dampen the effect of the outlier:
# Yeojohnson feature
from scipy.stats import yeojohnson

# Transform the 'Oldpeak' column using the Yeo-Johnson transformation
transformed_data, lambda_values = yeojohnson(new_data['RestingBP'])

# Store the transformed data back to the DataFrame
new_data['RestingBP_transformed'] = transformed_data

# Show the updated DataFrame
print(new_data)
```

```python
# Then we use the Tukey's method to check for the transformation effect on the outliers

# Calculate the IQR for RestingBP
Q1 = new_data['RestingBP_transformed'].quantile(0.25)
Q3 = new_data['RestingBP_transformed'].quantile(0.75)
IQR = Q3 - Q1

# Define Tukey's Fences
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Find outliers
outliers = new_data[(new_data['RestingBP_transformed'] < lower_fence) | (new_data['RestingBP_transformed'] > upper_fence)]

# print("DataFrame:")
# print(new_data)
print("\nOutliers based on Tukey's Fences:")
print(outliers)
```

```python
# We can also transform the feature using the Log feature to dampen the effect of the outlier:
# Log feature
new_data["Log_Of_RestingBP"] = [np.log(x) for x in new_data["RestingBP"]]
# Show data
new_data
```

```python
#Tukey's method

# Calculate the IQR for RestingBP
Q1 = new_data['Log_Of_RestingBP'].quantile(0.25)
Q3 = new_data['Log_Of_RestingBP'].quantile(0.75)
IQR = Q3 - Q1

# Define Tukey's Fences
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Find outliers
outliers = new_data[(new_data['Log_Of_RestingBP'] < lower_fence) | (new_data['Log_Of_RestingBP'] > upper_fence)]

# print("DataFrame:")
# print(new_data)
print("\nOutliers based on Tukey's Fences:")
print(outliers)
```

NOTE: For the outliers transformation, the yeojohnson transformation method was utilized because it can transform even zero and negative numbers unlike the log transformation feature which can only transform positive number. In the case if the dataset used in this analysis there are zero values and also negative number values and because of this, the yeojohnson method was used.

```python
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

For the model training and testing, we have to split the dataset into train-test. The train dataset will be used to train the machine learning models while the test dataset will be used to evaluate the performance of the model. In the case of this project, 90% of the dataset is used for training the model while 10% of the dataset is used for the test.

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_data, y_data, random_state=1, train_size=0.9)
```

Because of outlier detected ealier and to deal with them further, there is a need to rescale the data. We can achieve this through the **StandardScalar()** function.

**StandardScaler** is a preprocessing technique used to standardize (or scale) the features of a dataset. Standardization involves transforming the data in such a way that it has a mean of 0 and a standard deviation of 1. This process is also known as z-score normalization.


```python
scaler = StandardScaler()
# Transform the feature
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.transform(Xtest)
# Show feature
Xtrain
```

Here, we created a function which will allow us run some Machine Learning Models and then we can choose the best model. The following are the models are considered in this project.
1. Logistic Regression
2. KNeighbours Classifier
3. Decision Tree Classsifier
4. Random Forest Classifier

After this, the train accuracy and test accuracy were obtained. **But note :** The test accuracy rightly validates the model and not the train accuracy and so in this project, the best model will be selected based on the test accuracy.

```python
def models(Xtrain,ytrain):

      # Using Logistic Regression Algorithm for the Training Set
      from sklearn.linear_model import LogisticRegression
      log = LogisticRegression(random_state = 0)
      log.fit(Xtrain, ytrain)

      # Using KNeighborsClassifier Algorithm for the Training Set
      from sklearn.neighbors import KNeighborsClassifier
      knn = KNeighborsClassifier()
      knn.fit(Xtrain, ytrain)

      # Using DecisionTreeClassifier of tree class Algorithm for the Training Set
      from sklearn.tree import DecisionTreeClassifier
      tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
      tree.fit(Xtrain, ytrain)

      # Using RandomForestClassifier method of ensemble class Algorithm for the Training Set
      from sklearn.ensemble import RandomForestClassifier
      forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
      forest.fit(Xtrain, ytrain)

      # print model accuracy on the training data.
      print('[0]Logistic Regression Training Accuracy:', log.score(Xtrain, ytrain))
      print('[1]K Nearest Neighbor Training Accuracy:', knn.score(Xtrain, ytrain))
      print('[2]Decision Tree Classifier Training Accuracy:', tree.score(Xtrain, ytrain))
      print('[3]Random Forest Classifier Training Accuracy:', forest.score(Xtrain, ytrain))

      return log, knn, tree, forest
```

```python
# Get and train all of the models
model = models(Xtrain, ytrain)
```

```python
# Let's make prediction on the test data and get the confusion matrix, and test accuracy of each model
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    cm = confusion_matrix(ytest, model[i].predict(scaler.transform(Xtest)))
    #extracting TN, FP, FN, TP
    TN, FP, FN, TP = confusion_matrix(ytest, model[i].predict(scaler.transform(Xtest))).ravel()
    print(model[i])
    print("Confusion matrix \n", cm)
    print('Testing Accuracy = "{} !"'.format((TP + TN) / (TP + TN + FN + FP)))
    print()# Print a new line
```

```python
# We will choose Linear Regression, predict on the test data and create a classification report
LRmodel = model[0]
y_pred = LRmodel.predict(scaler.transform(Xtest))
print(classification_report(ytest, y_pred, target_names=["Negative", "Positive"]))
```

# Support Vector Machine

**Support Vector Machine (SVM)** is a supervised machine learning algorithm used for classification **(SVC)** and regression (SVR) tasks. It is a powerful and versatile algorithm that can be used for both linear and non-linear classification and regression tasks. SVMs are particularly well-suited for binary classification problems but can be extended to handle multi-class classification as well.

For the SVM in this project, Both the Support vector Classification and Support vector regression models were considered.

## Supprot Vector Classification (SVC)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Create an SVM classifier (SVC)
svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)

# Train the SVM classifier on the training data
svm_classifier.fit(Xtrain, ytrain)
```

```python
print('Support Vector Classifier Train Accuracy:', svm_classifier.score(Xtrain, ytrain))
```

```python
# Make predictions on the test data
y_pred = svm_classifier.predict(Xtest)

# Calculate the accuracy of the SVM model
accuracy = accuracy_score(ytest, y_pred)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

## Supprot Vector Regression (SVR)

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

# Create an SVR model
svm_regressor = SVR(kernel='linear', C=1.0, epsilon=0.2)

# Train the SVR model on the training data
svm_regressor.fit(Xtrain, ytrain)


```

```python
print('Support Vector Regressor Training Accuracy:', svm_regressor.score(Xtrain, ytrain))
```

Root Mean Squared Error (RMSE) is commonly used as a metric to evaluate the test accuracy of a Support Vector Regression (SVR) model, especially in regression tasks. RMSE measures the average magnitude of the errors between the predicted values and the actual (ground truth) values.
RMSE is a popular choice for evaluating SVR models because of its interpretability, sensitivity to errors, and close relationship with the loss function used during training. It provides a straightforward way to assess how well the SVR model's predictions match the actual values in a regression task.

```python
# Make predictions on the test data
y_pred = svm_regressor.predict(Xtest)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(ytest, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```
